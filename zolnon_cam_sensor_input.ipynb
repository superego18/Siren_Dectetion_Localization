```python

import time
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
```

```python
# Load the fine-tuned model
fine_tuned_model = load_model('best_fine_tuned_resnet50.h5', compile=False)

# Function to preprocess input frame and get predictions
def predict_frame(frame):
    # Resize the frame to match the input size of the model
    img = cv2.resize(frame, (224, 224))
    
    # Preprocess the image for model prediction
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    # Get model predictions
    predictions = fine_tuned_model.predict(img_array)

    # Interpret the model's predictions (assuming binary classification)
    class_labels = ['Ambulance', 'Not Ambulance']  # Replace with your actual class labels
    predicted_class = class_labels[int(predictions[0, 0] > 0.05)]

    return predicted_class

# Open a connection to the webcam (usually 0 or 1, depending on your setup)
cap = cv2.VideoCapture(0)
# Variables to track the time ambulance is detected
start_time = None
ambulance_detected = False
```

```python
with open('cam_output.txt', 'w') as file:
    # 0 값을 파일에 쓰기
    file.write('0')
```

```python
while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    # Perform prediction on the frame
    predicted_class = predict_frame(frame)

    # Display the frame with the predicted class
    cv2.putText(frame, f"Predicted class: {predicted_class}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Webcam Classification', frame)

    # Check if the predicted class is 'Ambulance'
    if predicted_class == 'Ambulance':
        if not ambulance_detected:
            # Start tracking time when ambulance is first detected
            start_time = time.time()
            ambulance_detected = True
        else:
            # Check if 3 seconds have passed
            elapsed_time = time.time() - start_time
            if elapsed_time >= 3:
                # Send 'vision' value 1 to Arduino
                with open('cam_output.txt', 'w') as file:
                    # 0 값을 파일에 쓰기
                    file.write('1')
                print('1111111111111111111111')
                ambulance_detected = False  # Reset detection after sending signal
    else:
        # Reset variables if predicted class is not 'Ambulance'
        start_time = None
        ambulance_detected = False

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam, close the window, and close serial communication
cap.release()
cv2.destroyAllWindows()
ser.close()
```
